### Write a function named palindrome_word(). It takes in a string as the only parameter. It returns True if the string is a palindrome by words. Otherwise, it returns False. A sample run of one possible implementation is shown below.


```python
def palindrome_word(a_str):
    special_symbols=[" ",",","?","!","'",'"',"`",".","-","_"]
    list=[]
    for char in a_str.split(" "): #split the sentences by word
      newChar='' 
      for c in char: #remove special symbols from the word and convert to lower case
        if c not in special_symbols:
          newChar += c.lower()
      list.append(newChar) #save the cleaned word to the list

    if list == list[::-1]: #check if the reversed list and actual list is same
      return True
    else:
      return False

my_string = 'King, are you glad you are king?'
print(palindrome_word(my_string))
```
