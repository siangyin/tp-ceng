### A program is required to record and display the results of a swimming competition. The swimming pool has 5 lanes. The swimmer name for each lane is first recorded. After the race, the timings for the swimmers are captured. The results for the race is then displayed.


### a.	Write a function inputSwimmers() that prompts for 5 swimmers' names and returns the names in a list. An example of the input session is as follows: ( 3 marks)
#### Enter lane 1 swimmer: Jin Jia Ho
#### Enter lane 2 swimmer: Mai Siao Siao
#### …
#### Enter lane 5 swimmer: Ho Seh Lah



```python
def inputSwimmers():

  namelist = []
  while len(namelist) < 5:
    userInput = input(f'Enter lane {len(namelist)+1} swimmer: ')
    if userInput == '':
      print('\nError: Swimmer name cannot be blank. Try again.\n')
      userInput = input(f'Enter lane {len(namelist)+1} swimmer: ')
    namelist.append(userInput)
  return namelist

swimmers = inputSwimmers()
```

### b.	Write another function inputTiming(swimmers) that has the list of swimmers names as paramter. The function creates an empty list, prompts and stores the timing for each of the swimmers. The timing list is returned. An example of the input session is as follows: (3 marks)
#### Enter timing for Jin Jia Ho: 47.15
#### Enter timing for Mai Siao Siao: 46.91
#### ….
#### Enter timing for Ho Seh Lah: 48.01


```python
def inputTiming(list):
  timingList = []
  for pax in list:
      timingList.append(float(input(f'Enter timing for {pax}: ')))
  return timingList

timing = inputTiming(swimmers)
```

```python
print(swimmers)
for name in swimmers:
    print(name)
```

### c.	Write another function printResults(swimmers, timings) that has the name and timing lists as parameters. The function prints a summary of the results as follows: (3 marks)
#### Jin Jia Ho           47.15s
#### Mai Siao Siao     46.91s
#### …
#### Ho Seh Lah        48.01s
#### The Fastest timing is 46.91s


```python
def printResults(name, time):
  i = 0
  while i < 5:
    print(f'{name[i]} {time[i]}s')
    i += 1
  print(f'\nThe Fastest timing is {min(time)}s')


printResults(swimmers, timing)
```

### d.	Modify the function in part (c) and name it printResults2(swimmers, timings) that prints the summary in sorted ascending order of the timing. Therefore, the summary may look like this: (5 marks)
#### Mai Siao Siao	46.91s
#### Jin Jia Ho	 47.15s
#### …
#### Ho Seh Lah	48.01s

```python
def printResults2(name, time):

  nestedList = []
  i = 0
  while i < len(name):
    nestedList.append([name[i], time[i]])
    i += 1

  def getTime(elem):
    return elem[1]

  nestedList.sort(key=getTime)
  for elem in nestedList:
    print(f'{elem[0]} {elem[1]}s ')

printResults2(swimmers, timing)
```
